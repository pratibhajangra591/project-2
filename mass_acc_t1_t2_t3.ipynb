```python
# Accretion of matter and radiation around the PBH from RD to MRE.

from scipy.integrate import quad
import numpy as np
from scipy.integrate import odeint
from matplotlib import rcParams
import matplotlib.pyplot as plt
from matplotlib import rcParams
import scipy.integrate as spi
from scipy import integrate
from accretion import*
import math
plt.rcParams.update({
  "text.usetex": True,
  "font.family": "serif"
})

bin_centres = np.geomspace(1e5, 1e-5, 6)
bin_edges = np.sqrt(bin_centres[:-1]*bin_centres[1:])
ratio = bin_edges[1]/bin_edges[0] 
bin_edges = np.append(bin_edges[0]/ratio, bin_edges) 
bin_edges = np.append(bin_edges, bin_edges[-1]*ratio) 
deltas = np.diff(bin_edges)
Δ_1_list = deltas.tolist()
Δ_2_list = deltas.tolist()
m_1 = np.geomspace(1e5, 1e-5, 6)
m_2 = np.geomspace(1e5, 1e-5, 6)
m_1_list = m_1.tolist()
m_2_list = m_2.tolist() 



plt.figure(figsize = (6, 5))
plt.loglog(m_1 , z_pbh(m_1, γ = 0.2), 'r')
plt.xlabel('$m_{pbh}$', fontsize = 29, labelpad = 10) 
plt.ylabel('$z(m_{pbh})$', fontsize = 29, labelpad = 10)
plt.xticks(fontsize = 26)
plt.yticks(fontsize = 26)
plt.title('$\gamma = 0.2$', fontsize = 24, pad = 8)
plt.show()
```

```python
# Accretion around the PBH from RD to MRE.
# Here, s = a/a_eq
# Here, v = c_s with c_s speed of sound in the surrounding medium.
# Here, γ is the ratio between the PBH mass and the horizon mass.





# Radiation accretion  with ρ = ρ_r0 (1 + z)⁴ up to z_eq.
# Using dt/dz as per t_1 = np.sqrt(3/(4 * np.pi * G * ρ_eq)) * ((s**2)/2) 
# (valid for small s as an approximation of Eq.(2.5) of https://arxiv.org/pdf/2304.05892.pdf
def m_racc_t1(γ, m, z):
    z_i = z_pbh(m,  γ)
    c_s  = c/np.sqrt(3)
    α_t1 = ((4 * π * (G**2))/(c_s**3)) * ((3/(4 * π * G * ρ_eq))**(1/2))
    z_i = z_pbh(m,  γ)
    return 1/( α_t1 *  (ρ_r0/2) * ((1 + z_eq)**2) * (z - z_i) * (z + z_i + 2) *  η_acc   + 1/m )




# Baryon accretion with ρ = ρ_m0 (1 + z)³ up to z_eq.
# Using dt/dz as per t_1 = np.sqrt(3/(4 * np.pi * G * ρ_eq)) * ((s**2)/2) 
# (valid for small s as an approximation of Eq.(2.5) of https://arxiv.org/pdf/2304.05892.pdf
def m_bacc_t1(γ, m, z):
    z_i = z_pbh(m,  γ)
    c_s  = c/np.sqrt(3)
    α_t1  = ((4 * π * (G**2))/(c_s**3)) * ((3/(4 * π * ρ_eq * G))**(1/2))
    z_i = z_pbh(m,  γ)
    return 1/( α_t1 * ρ_m0 * ((1 + z_eq)**2) * (z - z_i)  *  η_acc   + 1/m )




# Radiation accretion with ρ = ρ_r0 (1 + z)⁴ up to z_eq.
# Using dt/dz as per as per t_2 = t_eq * ((a/a_eq)**2).
def m_racc_t2(γ, m, z):
    z_i = z_pbh(m,  γ)
    c_s  = c/np.sqrt(3)
    α_t2 = ((4 * π * (G**2))/(c_s**3)) *  2 * t_eq
    z_i = z_pbh(m,  γ)
    return 1/( α_t2 *  (ρ_r0/2) * ((1 + z_eq)**2) * (z - z_i) * (z + z_i + 2) *  η_acc   + 1/m )




# Baryon accretion with ρ = ρ_m0 (1 + z)³ up to z_eq.
# Using dt/dz as per as per t_2 = t_eq * ((a/a_eq)**2).
def m_bacc_t2(γ, m, z):
    z_i = z_pbh(m,  γ)
    c_s  = c/np.sqrt(3)
    α_t2 = ((4 * π * (G**2))/(c_s**3)) *  2 * t_eq
    z_i = z_pbh(m,  γ)
    return 1/(α_t2 * ρ_m0 * ((1 + z_eq)**2) * (z - z_i)  *  η_acc   + 1/m )




#Accretion of radiation as per ρ = ρ_r0 (1 + z)⁴ up to z_eq.
# Using dt/dz as per t_3 = np.sqrt(3/(4 * np.pi * G * ρ_eq)) * ((2/3) * (s-2) * np.sqrt(s+1) + (4/3)) 
def m_racc_t3(γ, m, z):
    z_i = z_pbh(m,  γ)
    c_s  = c/np.sqrt(3)
    α_t3 = ((4 * π * (G**2))/(c_s**3)) * ((3/(8 * π * G * ρ_c0))**(1/2))
    def integral(z):
        y = (1 + z)
        a = Ω_r0
        b = Ω_m0
        first_term = np.sqrt(a) * (y**3) * (2 * a**2 * y**2 - a * b * y - 3 * b**2)
        second_term = 3 * (b**(5/2)) * (y**(5/2)) * np.sqrt((a * y/b) + 1) * np.arcsinh(np.sqrt(a * y/b)) 
        third_term = 4 * (a**(5/2)) * np.sqrt((y**(5)) * (a * y + b)) 
        return (first_term + second_term)/ third_term 
    return 1/(α_t3 *  ρ_r0  * (integral(z) - integral(z_i)) *  η_acc   + 1/m )





### Accretion of baryons as per ρ = ρ_r0 (1 + z)³ up to z_eq.
# Using dt/dz as per t_3 = np.sqrt(3/(4 * np.pi * G * ρ_eq)) * ((2/3) * (s-2) * np.sqrt(s+1) + (4/3)) 
def m_bacc_t3(γ, m, z):
    z_i = z_pbh(m,  γ)
    c_s  = c/np.sqrt(3)
    α_t3 = ((4 * π * (G**2))/(c_s**3)) * ((3/(8 * π * ρ_c0 * G))**(1/2))
    def integral(z):
        y = (1 + z)
        a = Ω_r0
        b = Ω_m0
        first_term = np.sqrt(a) * (y**3) * (a * y + b)
        second_term = (b**(3/2)) * (y**(5/2)) * np.sqrt((a* y/b) + 1) * np.arcsinh(np.sqrt(a * y/b)) 
        third_term = (a**(3/2)) * np.sqrt((y**(5)) * (a * y + b)) 
        return (first_term - second_term)/ third_term 
    return 1/(α_t3 *  ρ_m0  * (integral(z) - integral(z_i)) *  η_acc  + 1/m )
```

```python
# Radiation and matter accretion from RD to upto MRE for different values of γ.
# Here, γ is the ratio between the PBH mass and the horizon mass.
# Using t_i(m_i) = np.sqrt(3/(4 * np.pi * G * ρ_eq)) * ((s**2)/2).





# Using dt/dz as per t_1 = np.sqrt(3/(4 * np.pi * G * ρ_eq)) * ((s**2)/2) 
# (valid for small s as an approximation of Eq.(2.5) of https://arxiv.org/pdf/2304.05892.pdf
def Horizon_by_PBH_mass1(γ, m, z, radiation):
    if radiation == True:
        value = (m_racc_t1(γ, m, z) - m)/m
    else:
        value = (m_bacc_t1(γ, m, z) - m)/m
    return value



# Using dt/dz as per as per t_2 = t_eq * ((a/a_eq)**2).
def Horizon_by_PBH_mass2(γ, m, z, radiation): 
    if radiation == True:
        value = (m_racc_t2(γ, m, z) - m)/m
    else:
        value = (m_bacc_t2(γ, m, z) - m)/m
    return value


# Using dt/dz as per t_3 = np.sqrt(3/(4 * np.pi * G * ρ_eq)) * ((2/3) * (s-2) * np.sqrt(s+1) + (4/3)) 
def Horizon_by_PBH_mass3(γ, m, z, radiation): 
    if radiation == True:
        value = (m_racc_t3(γ, m, z) - m)/m
    else:
        value = (m_bacc_t3(γ, m, z) - m)/m
    return value



γ = np.geomspace(1, 0.1, 5)
```
