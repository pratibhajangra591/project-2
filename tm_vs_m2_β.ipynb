```python
# Final merger time for the reference systems with masses m_1 = 1e5 M_solar
# and m_2 = 1 M_solar, as a function of the initial 
# angular momentum j0 for binaries having Power Law PDF (1-1e5) for PBHs having alpha-disk 
# with dynamical friction (Ostriker) as dissipative force.

%matplotlib inline
import numpy as np
from scipy.integrate import odeint
import math
from matplotlib import rcParams
import matplotlib.pyplot as plt
from matplotlib.ticker import LogFormatterExponent
from matplotlib.colors import LogNorm
from matplotlib import ticker, cm
from accretion import *
import matplotlib as mpl
from matplotlib.lines import Line2D
from scipy.interpolate import interp1d
from scipy.integrate import quad
from scipy.special import gamma
from imripy import halo, constants as c, merger_system as ms, inspiral, waveform, detector, plot_utils
from imripy.inspiral import forces
import imripy.constants as c

plt.rcParams.update({
  "text.usetex": True,
  "font.family": "serif"
})
solar_mass_to_pc = 4.8e-14
year_to_pc = 0.3064



π = np.pi
Ω_cdm = 0.85
G = 4.4959e-15            #in units of M☉^-1 pc^3 yr^-2
c = 0.3068                #in units of pc yr^-1
ρ_eq = 3.1808e3           #in units of M☉ pc^-3 with ρ_eq=2.1548e-16 kg m^-3
pc = 3.0857e16            # in meters
yr = 3.154e7              # in units of seconds
t_eq = 1.5923e12/yr      # in units of seconds
t_m = 13.78e9             #in units of yrs corresponding to t_0=13.78Gyr
t_0 = 13.78e9             #in units of yrs corresponding to t_0=13.78Gyr

σ_eq = 0.005
ρ_m = 4e19                #ρ_m = 4e19 M☉ Gpc^-3



#1.) SystemProp(m1, m2, halo, D=D, inclination_angle = 0., pericenter_angle=0.,
# baryonicHalo=None, includeHaloInTotalMass=False):
    
#2.) EvolutionOptions(accuracy=1e-10, verbose=1, elliptic=True, m2_change=False,
# dissipativeForces=None, gwEmissionLoss = True, dynamicalFrictionLoss = True,
# considerRelativeVelocities=False, progradeRotation = True, **kwargs)

#3.) Evolve(sp, a_0, e_0=0., a_fin=0., t_0=0., t_fin=None, opt=EvolutionOptions())
# t_fin = None means t_fin = Time of coalescence on the system.


 
#4.) opt_0 = EvolutionOptions(accuracy=1e-10, verbose=1, elliptic=True, m2_change=False,
# dissipativeForces=None, gwEmissionLoss = True, dynamicalFrictionLoss = True,
# considerRelativeVelocities=False, progradeRotation = True)or 
# or
#4.) opt_0 = inspiral.Classic.EvolutionOptions(dynamicalFrictionLoss=False, verbose=1, accuracy=1e-10)
```

```python
# Final merger time vs m1 for IMRI with β-accretion disk having dynamical 
# friction as the dissipative force.



# Setting initial and final conditions of the binary evolution for an eccentric orbit
e0 = 0.999
m1_arr = np.geomspace(1e5, 1e7, 4)  # in units of solar mass
m2_arr = np.geomspace(1, 100, 50)  # in units of solar mass


fig,ax = plt.subplots(1, figsize = (14, 7.8))
plt.rc('lines', linewidth = 1.7)
plt.rcParams['font.family'] = 'serif'
plt.tick_params(axis = 'both', right=True, top=True, which = 'minor', width= 1.2, length = 3, direction = 'in')
plt.tick_params(axis = 'both', right=True, top=True, which = 'major', width= 1.2, length = 7, direction = 'in')

colors = ['C0', 'C1', 'C2', 'C3']
plot_lines = []


for k, (m1, color) in enumerate(zip(m1_arr, colors)):
    tm_arr_0 = np.zeros(len(m2_arr))
    tm_arr_DF = np.zeros(len(m2_arr))
    tm_arr_gwDF = np.zeros(len(m2_arr))
    for i, m2 in enumerate(m2_arr):
        m1 =  m1_arr[k] * solar_mass_to_pc  
        m2 =  m2_arr[i] * solar_mass_to_pc  
        D = 0
        a0  =  500 * ((6 * G * 1e5)/(c**2)) #  a_i = (500 * r_isco of 1e5 solar mass) in units of parsec
        accuracy = 1e-10


        sp_0 = ms.SystemProp(m1, m2, halo.ConstHalo(0.), D)
        opt_0 = inspiral.Classic.EvolutionOptions(accuracy=accuracy, verbose=1, 
                    gwEmissionLoss = True, dynamicalFrictionLoss = False)
        R_fin_0 = sp_0.r_isco()      # The final condition for the evolution, Choosen equal to r_icso
        ev_0 = inspiral.Classic.Evolve(sp_0, a0, e0, a_fin=R_fin_0, opt=opt_0)   
        tm_arr_0[i] = ev_0.t[-1]/year_to_pc


        alpha = 0.1
        f_edd = 0.1
        eps   = 0.1
        betaDisk = halo.BetaDisk(m1, alpha, f_edd, eps)
        sp_beta = ms.SystemProp(m1, m2, halo.ConstHalo(0.), baryonicHalo = betaDisk, D=D)
        R_fin = sp_beta.r_isco()      # The final condition for the evolution, Choosen equal to r_icso

        opt_DF    = inspiral.Classic.EvolutionOptions(gwEmissionLoss = False, dissipativeForces={forces.GasDynamicalFriction()}, 
                    considerRelativeVelocities= True, progradeRotation=False, accuracy=accuracy, verbose=1)

        opt_gwDF    = inspiral.Classic.EvolutionOptions(dissipativeForces={forces.GWLoss(), forces.GasDynamicalFriction()}, 
                      gwEmissionLoss = True, dynamicalFrictionLoss = True, considerRelativeVelocities= True, 
                      progradeRotation = False, accuracy=accuracy, verbose=1)

        ev_beta_DF = inspiral.Classic.Evolve(sp_beta, a0, e0, a_fin=R_fin, opt=opt_DF)   
        tm_arr_DF[i] = ev_beta_DF.t[-1]/year_to_pc
        ev_beta_gwDF = inspiral.Classic.Evolve(sp_beta, a0, e0, a_fin=R_fin, opt=opt_gwDF)   
        tm_arr_gwDF[i] = ev_beta_gwDF.t[-1]/year_to_pc



    
    ax.loglog(m2_arr, tm_arr_DF,  color = color, linestyle = '--')
    ax.loglog(m2_arr, tm_arr_0,  color = color, linestyle = ':')
    ax.loglog(m2_arr, tm_arr_gwDF, color = color, linestyle = '-')
    
    
    slope_0, intercept_0 = np.polyfit(np.log(m2_arr), np.log(tm_arr_0), 1)
    slope_DF, intercept_DF = np.polyfit(np.log(m2_arr), np.log(tm_arr_DF), 1)
    slope_gwDF, intercept_gwDF = np.polyfit(np.log(m2_arr), np.log(tm_arr_gwDF), 1)
    print('\v')
    print("No Accretion (GW),", "intercept = ", '%0.4f'%intercept_0, "," , "m =", '%0.4f'%slope_0)
    print('β - disk(DF),', "intercept = ", '%0.4f'%intercept_DF,  "," , "m =", '%0.4f'%slope_DF)
    print('β - disk (gw + DF) ,', "intercept = ", '%0.4f'%intercept_gwDF,  "," , "m =", '%0.4f'%slope_gwDF)

    
    

    
ax.tick_params(axis='x', pad = 8.9)
ax.tick_params(axis='y', pad = 8.9)
# In total 3x4 lines have been plotted
lines = ax.get_lines()
legend1 = plt.legend([lines[i] for i in [0,1,2]], ["DF", "GW", "DF + GW"], title = 'Dissiptaive Force',
        title_fontsize = 23, loc = 'upper right', prop = {"size":13})
legend2 = plt.legend([lines[i] for i in [2,5,8,11]], ['$10^{5}$', r'$4.64\times 10^{5}$', 
           r'$2.15\times 10^{6}$', '$10^{7}$'], title = '$m_{1} \: \mathrm{[M_{\odot}]}$',
           title_fontsize = 27, loc = (1.02, 0.25), prop = {"size":28})

ax.add_artist(legend1)
ax.add_artist(legend2)



plt.xlabel('$m_{2} \, \mathrm{[M_{\odot}]}$', fontsize = 35, labelpad = 15)
plt.ylabel('Final merger time  [yr]', fontsize = 35, labelpad = 15)
plt.xticks(fontsize = 30)
plt.yticks(fontsize = 30)
plt.title(r'$ \beta-\mathrm{disk},\, a_{i} = 500\,r_\mathrm{isco}(m = 10^{5}\, \mathrm{M_{\odot}}),\, e_{i} = 0.999$', fontsize = 32, pad = 9)
plt.savefig("tm_vs_m2_β_DF.pdf", bbox_inches = 'tight')
plt.show()   
```

```python
# Final merger time vs m1 for IMRI with β-accretion disk having Type-I
# migration as the dissipative force.
# Type- I migration is not valid for e > 0.001.



# Setting initial and final conditions of the binary evolution for an eccentric orbit
e0 = 1e-4   #Type- I migration is not valid for e > 0.001.
m1_arr = np.geomspace(1e5, 1e7, 4)  # in units of solar mass
m2_arr = np.geomspace(1, 100, 50)  # in units of solar mass


fig,ax = plt.subplots(1, figsize = (14, 7.8))
plt.rc('lines', linewidth = 1.7)
plt.rcParams['font.family'] = 'serif'
plt.tick_params(axis = 'both', right=True, top=True, which = 'minor', width= 1.2, length = 3, direction = 'in')
plt.tick_params(axis = 'both', right=True, top=True, which = 'major', width= 1.2, length = 7, direction = 'in')

colors = ['C0', 'C1', 'C2', 'C3']
plot_lines = []
for k, (m1, color) in enumerate(zip(m1_arr, colors)):
    tm_arr_0 = np.zeros(len(m2_arr))
    tm_arr_TypeI = np.zeros(len(m2_arr))
    tm_arr_gwTypeI = np.zeros(len(m2_arr))
    for i, m2 in enumerate(m2_arr):
        m1 =  m1_arr[k] * solar_mass_to_pc  
        m2 =  m2_arr[i] * solar_mass_to_pc  
        D = 0
        a0  =  500 * ((6 * G * 1e5)/(c**2)) #  a_i = (500 * r_isco of 1e5 solar mass) in units of parsec
        accuracy = 1e-10
        
        
        sp_0 = ms.SystemProp(m1, m2, halo.ConstHalo(0.), D)
        opt_0 = inspiral.Classic.EvolutionOptions(accuracy=accuracy, verbose=1, 
                    gwEmissionLoss = True, dynamicalFrictionLoss = False)
        R_fin_0 = sp_0.r_isco()      # The final condition for the evolution, Choosen equal to r_icso
        ev_0 = inspiral.Classic.Evolve(sp_0, a0, e0, a_fin=R_fin_0, opt=opt_0)   
        tm_arr_0[i] = ev_0.t[-1]/year_to_pc
        
        
        alpha = 0.1
        f_edd = 0.1
        eps   = 0.1
        betaDisk = halo.BetaDisk(m1, alpha, f_edd, eps)
        sp_beta = ms.SystemProp(m1, m2, halo.ConstHalo(0.), baryonicHalo = betaDisk, D=D)
        R_fin = sp_beta.r_isco()      # The final condition for the evolution, Choosen equal to r_icso
       
        opt_TypeI    = inspiral.Classic.EvolutionOptions(gwEmissionLoss = False, dissipativeForces={forces.GasInteraction()}, 
                    considerRelativeVelocities= False, progradeRotation= True, accuracy=accuracy, verbose=1)
        
        opt_gwTypeI  = inspiral.Classic.EvolutionOptions(dissipativeForces={forces.GWLoss(), forces.GasInteraction()}, 
                      gwEmissionLoss = True, dynamicalFrictionLoss = True, considerRelativeVelocities= False, 
                      progradeRotation = True, accuracy=accuracy, verbose=1)
       
        ev_beta_TypeI = inspiral.Classic.Evolve(sp_beta, a0, e0, a_fin=R_fin, opt=opt_TypeI)   
        tm_arr_TypeI[i] = ev_beta_TypeI.t[-1]/year_to_pc
        ev_beta_gwTypeI = inspiral.Classic.Evolve(sp_beta, a0, e0, a_fin=R_fin, opt=opt_gwTypeI)   
        tm_arr_gwTypeI[i] = ev_beta_gwTypeI.t[-1]/year_to_pc
        
    
    ax.loglog(m2_arr, tm_arr_TypeI, color = color, linestyle = '--')
    ax.loglog(m2_arr, tm_arr_0, color = color, linestyle = ':')
    ax.loglog(m2_arr, tm_arr_gwTypeI, color = color, linestyle = '-')

    
    slope_0, intercept_0 = np.polyfit(np.log(m2_arr), np.log(tm_arr_0), 1)
    slope_TypeI, intercept_TypeI = np.polyfit(np.log(m2_arr), np.log(tm_arr_TypeI), 1)
    slope_gwTypeI, intercept_gwTypeI = np.polyfit(np.log(m2_arr), np.log(tm_arr_gwTypeI), 1)
    print('\v')
    print("No Accretion (GW),", "intercept = ", '%0.4f'%intercept_0, "," , "m =", '%0.4f'%slope_0)
    print('β - disk(Type-I)', "intercept = ", '%0.4f'%intercept_TypeI,  "," , "m =", '%0.4f'%slope_TypeI)
    print('β - disk (gw + Type-I) ', "intercept = ", '%0.4f'%intercept_gwTypeI,  "," , "m =", '%0.4f'%slope_gwTypeI)

    
    
    
ax.tick_params(axis='x', pad = 8.9)
ax.tick_params(axis='y', pad = 8.9)
# In total 3x4 lines have been plotted
lines = ax.get_lines()
legend1 = plt.legend([lines[i] for i in [0,1,2]], ["Type - I", "GW", "Type - I + GW"], title = 'Dissiptaive Force',
        title_fontsize = 23, loc = 'upper right', prop = {"size":13})
legend2 = plt.legend([lines[i] for i in [2,5,8,11]], ['$10^{5}$', r'$4.64\times 10^{5}$', 
           r'$2.15\times 10^{6}$', '$10^{7}$'], title = '$m_{1} \: \mathrm{[M_{\odot}]}$',
           title_fontsize = 27, loc = (1.02, 0.25), prop = {"size":28})

ax.add_artist(legend1)
ax.add_artist(legend2)

plt.xlabel('$m_{2} \, \mathrm{[M_{\odot}]}$', fontsize = 35, labelpad = 15)
plt.ylabel('Final merger time  [yr]', fontsize = 35, labelpad = 15)
plt.xticks(fontsize = 30)
plt.yticks(fontsize = 30)
plt.title(r'$ \beta-\mathrm{disk},\, a_{i} = 500\,r_\mathrm{isco}(m = 10^{5}\, \mathrm{M_{\odot}}),\, e_{i} = 10^{-4}$', fontsize = 32, pad = 9)
plt.savefig("tm_vs_m2_β_TypeI.pdf", bbox_inches = 'tight')
#plt.grid()
plt.show()   
```
