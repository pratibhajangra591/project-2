```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate
import  vrel
from accretion import*


plt.loglog(vrel.z_arr, vrel.c_s_arr, 'C0')
plt.ylabel(r"$c_{s}\, \mathrm{[km \,s^{-1}}]$ ", fontsize = 23)
plt.xlabel(r"$z+1$", fontsize = 23)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.title('Digitized with $ 10\leq z \leq 1100$', fontsize = 20, pad = 7)
plt.show()
```

```python
plt.figure(figsize = (12, 5))

x = vrel.z_arr
y = vrel.c_s_arr
x  = (x[0], x[1], x[2] )
y  = (y[0], y[1], y[2])

plt.loglog(x, y, 'gd',  label = 'data')

p = np.polyfit(np.log(x), np.log(y), 1)  # F(x) = (x**m)(10**b) with m = slope, b = intercept.
z = np.polyval(p, np.log(x))
plt.loglog(x, np.exp(z), 'r',  label = 'fit')
print("Slope and intercept of fitting function =", p) 
plt.ylabel(r"$c_{s}\, \mathrm{[km \,s^{-1}}]$ ", fontsize = 23)
plt.xlabel(r"$z+1$", fontsize = 23)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.show()

# So, the fitting function is c_s(z<10) = (10**(-4.1193/np.log(10)))  * (z**0.9823)  
```

```python
def c_s_extrapolation(variable):
    if variable < 10:
        m = 0.9823
        b = -4.1193/np.log(10)
        return   (variable**m) * (10**b)
    else:
        x = vrel.z_arr
        y = vrel.c_s_arr
        f = interpolate.interp1d(x, y , fill_value = "extrapolate")
        return f(variable)


z_arr = np.geomspace(1, 10000, 100000)
cs_arr = np.zeros(len(z_arr))
for i, z in enumerate(z_arr):
    cs_arr[i] = c_s_extrapolation(z)
plt.loglog(z_arr, cs_arr, 'C0')   
plt.ylabel(r"$c_{s}\, \mathrm{[km \,s^{-1}}]$ ", fontsize = 23)
plt.xlabel(r"$z+1$", fontsize = 23)
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
plt.title('Digitized with extapolation for $z < 10$', fontsize = 20, pad = 7)
plt.show()
```

```python
def c_s(z):
     = 1.72
    z_dec = 130
    return  5.74 * np.sqrt((1+z)/1000) * (((((1+z_dec)/(1 + z))**) + 1)**(-1/(2 * ))) 

def v_rel_digitized(z):
    return  vrel_extrapolation(z) 


def Mach_number(z):
        return v_rel_digitized(z)/c_s(z)

def v_eff_A(z):
    if Mach_number(z) > 1:
        return  c_s(z) * (((16/np.sqrt(2 * np.pi)) * (Mach_number(z)**3))**(1/6))  
    else:
        return  c_s(z) * np.sqrt(1 + (Mach_number(z)**2))


def v_eff_B(z):
    e = 2.71828
    if Mach_number(z) > 1:
        return  c_s(z) * Mach_number(z) * ((np.sqrt(2/np.pi) * np.log((2/e) * Mach_number(z)))**(-1/3))  
    else:
        return  c_s(z) * np.sqrt(1 + (Mach_number(z)**2))



def c_s_digitized(z):
    return  c_s_extrapolation(z)


def v_effA_digitized(z):
    return  veffA_extrapolation(z) 


def v_effB_digitized(z):
    return  veffB_extrapolation(z) 



colors = ['C0', 'C1','C2','C3','C4', 'C5']
fig,ax = plt.subplots(1, figsize = (8, 6))
plt.rc('lines', linewidth = 1.7)
plt.rcParams['font.family'] = 'serif'
plt.tick_params(axis = 'both', right=True, top=True, which = 'minor', width= 1.2, length = 4, direction = 'in')
plt.tick_params(axis = 'both', right=True, top=True, which = 'major', width= 1.2, length = 9, direction = 'in')


z_array = np.geomspace(1, 10000, 100000)
cs_array = np.zeros(len(z_array))
v_effA_array = np.zeros(len(z_array))
v_effB_array = np.zeros(len(z_array))

cs_digitized_array = np.zeros(len(z_array))
v_effA_digitized_array = np.zeros(len(z_array))
v_effB_digitized_array = np.zeros(len(z_array))

v_rel_analytic_array = np.zeros(len(z_array))

for i, z in enumerate(z_array):
    cs_array[i] = c_s(z) 
    v_effA_array[i] = v_eff_A(z) 
    v_effB_array[i] = v_eff_B(z) 
    cs_digitized_array[i] = c_s_digitized(z)
    v_effA_digitized_array[i] = v_effA_digitized(z)
    v_effB_digitized_array[i] = v_effB_digitized(z)

    v_rel_analytic_array[i] = v_rel_digitized(z)
    

ax.loglog(z_array, v_effB_array, 'C2')
ax.loglog(z_array, v_effA_array, 'C1')
ax.loglog(z_array, cs_array, 'C0')

ax.loglog(z_array, v_effB_digitized_array, 'C2', ls = '--')
ax.loglog(z_array, v_effA_digitized_array, 'C1', ls = '--')
ax.loglog(z_array, cs_digitized_array, 'C0', ls = '--')

ax.loglog(z_array, v_rel_analytic_array, 'k', ls = '-.') 
ax.tick_params(axis='x', pad = 9.9)
ax.tick_params(axis='y', pad = 9.9)


plt.ylim(0.5, 10)
plt.xlim(10, 1700)

lines = ax.get_lines()
legend1 = plt.legend([lines[i] for i in [0, 1, 2]], [ r'$v_\mathrm{eff, \, B} $', r'$v_\mathrm{eff, \, A}$', '$c_{s}$'],
           title_fontsize = 12, loc = 'lower right', prop = {"size":18.5})
legend2 = plt.legend([lines[i] for i in [0, 3]], ['Analytic', r'Digitized'],
           title_fontsize = 12, loc = (0.74, 0.33), prop = {"size":15})

ax.add_artist(legend1)
ax.add_artist(legend2)

plt.text(12, 2.8, '$v_\mathrm{rel,\, digitized}$', fontsize = 24, rotation = 30)

plt.ylabel(r"Velocity$\, \mathrm{[km \,s^{-1}}]$ ", fontsize = 24, labelpad = 8)
plt.xlabel(r"$z + 1$", fontsize = 24, labelpad = 8)
plt.xticks(fontsize = 21)
plt.yticks(fontsize = 21)
plt.savefig('velocity_analytic_vs_digitized_extrapolated.pdf', bbox_inches = 'tight')
plt.show()
```
