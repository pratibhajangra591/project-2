```python
# Change in the masses of the PBHs due to accretion in radiation domination to 
# z_eq, post PBH formation.



%matplotlib inline
import numpy as np
from scipy.integrate import odeint
import math
from matplotlib import rcParams
import matplotlib.pyplot as plt
from matplotlib import rcParams
from module2 import *
plt.rcParams.update({
  "text.usetex": True,
  "font.family": "serif"
})
 

π = np.pi
Ω_cdm = 0.85
G = 6.67e-11                # in units of m^3⋅kg^−1⋅s^−2
M_solar = 1.989e30          # in units of kg
A = 1.495e11                # Astronomical unit
pc = 3.085e16               # in unist of meter. 
Λ = 2.036e-35               # in units of s⁻2
c = 3e8                     # in units of ms⁻¹
yr = 3.154e7                # in units of seconds
t_m = 13.78e9 * yr          #in units of yrs corresponding to t_0=13.78Gyr
t_eq = 1.5923e12            # in units of seconds
ρ_eq = 2.1548e-16           # in units of kg m^-3
ρ_meq = ρ_eq/2              # in units of kg m^-3
σ_eq = 0.005
a_eq = 2.9374e-4
M_H_eq = 3.1e16             #Horizon mass at z_eq in units of solar mass,



# In radiation domination, t as a function of redshift z when ρ = ρ_r0 * (a**-4).
def t(z):
    return (1/2) * (((8 * π * ρ_r0 * G)/3)**(-1/2)) * (1/((1+z)**2))


z = np.geomspace(1e6, z_eq, 100000)
z_list = z.tolist()
t_list = np.zeros(len(z_list))
for i, z in enumerate(z_list):
    t_list[i] = t(z)


print("z_eq =", '%0.2f'%z_eq)   
print('t(z_eq) =', '%0.4e'%t(z_eq), "s")    
fig, ax = plt.subplots(1, figsize = (8, 5))
plt.rc('lines', linewidth = 1.3)

plt.loglog(z_list, t_list, 'r')
plt.axvline(x = z_eq,  color = 'k',  linestyle = ':')
plt.xlim(1.5e6, 2e3)
plt.xticks(fontsize = 23)
plt.yticks(fontsize = 23)
plt.xlabel('$z$',fontsize = 29.5, labelpad = 12)
plt.ylabel('$t\, [s]$',fontsize = 29.5, labelpad = 12)
plt.title('Time at any redshift from RD to MRE', fontsize = 23, pad = 6)
plt.savefig('tvsz.pdf', bbox_inches = 'tight')
plt.show()
```

```python
# t as a function of redshift z when ρ = ρ_0 * ((a**-4) + (a**-3) + (constant))
# with values of ρ_c0, Ω_r0 and Ω_m0 used in our calculations above. 



t_1 = 2.1e12    # time at the end of radiation era
t_2 = 2.4e17    # time at the end of matter era
t_0 = t_m       # time at the age of the Universe today
H_0 = np.sqrt((8 * π * G * ρ_c0)/3)


def t_tabashi(z):
    C = (t_1**(-1/2)) * ((t_1/t_2)**(2/3)) * (np.exp(H_0 *(t_2 -t_0)))
    return (1/(C**2)) * ((1/(1+z))**2)



z = np.geomspace(1e6, z_eq, 100000)
z_list = z.tolist()
t_list = np.zeros(len(z_list))
for i, z in enumerate(z_list):
    t_list[i] = t_tabashi(z)

    

    
print("z_eq =", '%0.2f'%z_eq)   
print('t(z_eq) =', '%0.4e'%t_tabashi(z_eq), "s")
fig, ax = plt.subplots(1, figsize = (8, 5))
plt.rc('lines', linewidth = 1.3)

plt.loglog(z_list, t_list, 'm')
plt.axvline(x = z_eq,  color = 'k',  linestyle = ':')

plt.xlim(1.5e6, 2e3)
plt.xticks(fontsize = 23)
plt.yticks(fontsize = 23)
plt.xlabel('$z$',fontsize = 29.5, labelpad = 12)
plt.ylabel('$t\, [s]$',fontsize = 29.5, labelpad = 12)
plt.title('Time at any redshift from RD to MRE', fontsize = 23, pad = 6)
#plt.savefig('tvsz.pdf', bbox_inches = 'tight')
plt.show()
```
